```python
from gi_scraper import Scraper
import requests

sc = Scraper(headless=False, workers=2)
```

```python
# functions 
def fetch_urls(search_term, num_images):
    for query, count in {search_term: num_images}.items():
        print("Querying...", query)

        # scrape method returns a stream object
        stream = sc.scrape(query, count)

        # stream.get method yields Response object with following attributes
        # - query (str): The query associated with the response.
        # - name (str): The name attribute of the response.
        # - src_name (str): The source name attribute of the response.
        # - src_page (str): The source page attribute of the response.
        # - thumbnail (str): The thumbnail attribute of the response.
        # - image (str): The image attribute of the response.
        # - width (int): The width attribute of the response.
        # - height (int): The height attribute of the response.

        for response in stream.get():
            # response.to_dict returns python representable dictionary
            # print(response.to_dict())
            # print(response.image)
            if response.image not in images_url:
                images_url.append(response.image)
                print(len(images_url))
                
def save_images(images_url, path):
    idx = 0 
    for url in set(images_url): # ensure there are no duplicates 
        try: 
            r = requests.get(url) 
            if r.status_code == 200:
                idx += 1 
                filename = path + f"{idx:03d}.jpg"
                with open(filename, "wb") as f: 
                    f.write(r.content)
                    print(f"Writing to {filename}...")
        except: 
            print("Exception occurred")
            continue 
```

```python
common_to_scientific = {
    "COAL TIT":"PERIPARUS ATER"
}
```

```python
for species, scientific_name in common_to_scientific.items(): 
    images_url = []
    search_terms = [scientific_name, scientific_name + " PHOTOGRAPHY", scientific_name + " PICTURE", species, species + " BIRD", species + " BIRD PHOTOGRAPHY", species + " BIRD PICTURE", species + " WILDLIFE"] 
    term_index = 0
    num_images = 250
    while len(images_url) < num_images and term_index < len(search_terms): 
        fetch_urls(search_terms[term_index], num_images)
        term_index += 1
    save_images(images_url, path = f"{species}/")
```

```python
# terminate the query 
sc.terminate()
```
